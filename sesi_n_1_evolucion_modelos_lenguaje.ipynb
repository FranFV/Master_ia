## 1. Introducci√≥n y Contexto

### ¬øQu√© es el Procesamiento de Lenguaje Natural (NLP)?

El **Procesamiento de Lenguaje Natural** (Natural Language Processing, NLP) es el campo de la inteligencia artificial que permite a las m√°quinas:

- **Entender** el lenguaje humano (comprensi√≥n)
- **Generar** texto coherente y relevante (generaci√≥n)
- **Extraer** informaci√≥n significativa de textos
- **Traducir** entre idiomas
- **Analizar** sentimientos, intenciones y contexto

### ¬øPor qu√© es importante?

- El 80% de los datos empresariales son no estructurados (textos, emails, documentos)
- Permite automatizar tareas que antes requer√≠an comprensi√≥n humana
- Aplicaciones: chatbots, traducci√≥n autom√°tica, an√°lisis de redes sociales, asistentes virtuales
- Base de sistemas como ChatGPT, Google Translate, Siri, Alexa

---

### üìä Evoluci√≥n Hist√≥rica del NLP

#### **1. Era de las Reglas (1950-1990)**
- **Enfoque**: Gram√°ticas formales, sistemas basados en reglas escritas a mano
- **Ejemplo**: ELIZA (1966), primer chatbot con reglas if-then
- **Limitaciones**: No escalables, fr√°giles ante variaciones del lenguaje

#### **2. Era Estad√≠stica (1990-2010)**
- **Enfoque**: Modelos probabil√≠sticos, n-gramas, Bag of Words, TF-IDF
- **T√©cnicas**: Naive Bayes, HMM (Hidden Markov Models), CRF
- **Ventajas**: Aprenden de datos, m√°s robustos
- **Limitaciones**: No capturan sem√°ntica profunda ni contexto largo

#### **3. Era del Deep Learning (2010-2017)**
- **Enfoque**: Redes neuronales: RNN, LSTM, GRU
- **Hito**: Word2Vec (2013), GloVe (2014) - embeddings que capturan sem√°ntica
- **Ventajas**: Capturan contexto secuencial
- **Limitaciones**: Problema del gradiente desvaneciente, lentitud en secuencias largas

#### **4. Era de los Transformers (2017-2020)**
- **Hito**: Paper "Attention Is All You Need" (2017)
- **Modelos**: BERT (2018), GPT-2 (2019), RoBERTa, T5
- **Innovaci√≥n**: Mecanismo de atenci√≥n, procesamiento paralelo
- **Ventajas**: Capturan contexto bidireccional, escalan mejor

#### **5. Era de los LLM - Modelos Generativos (2020-actualidad)**
- **Modelos**: GPT-3/4 (OpenAI), LLaMA (Meta), Claude (Anthropic), Gemini (Google)
- **Caracter√≠sticas**:
  - Billones de par√°metros
  - Entrenamiento con billones de tokens
  - Capacidades emergentes (razonamiento, c√≥digo, multimodalidad)
- **T√©cnicas**: Prompt engineering, RAG, fine-tuning, agentes

---

## 2. Preprocesamiento B√°sico y An√°lisis Ling√º√≠stico (NLP Cl√°sico)

### Fundamentos del NLP Cl√°sico

Antes de los modelos modernos, el procesamiento de texto requer√≠a m√∫ltiples pasos de preprocesamiento:

#### **Tokenizaci√≥n**
Dividir el texto en unidades b√°sicas (palabras, oraciones)
- Ejemplo: `"Hola mundo"` ‚Üí `["Hola", "mundo"]`

#### **Stopwords (palabras vac√≠as)**
Eliminar palabras comunes sin significado relevante
- Ejemplo: "el", "la", "de", "que", "y"

#### **Stemming (derivaci√≥n)**
Reducir palabras a su ra√≠z eliminando sufijos
- Ejemplo: `"corriendo"` ‚Üí `"corr"`
- M√°s r√°pido pero menos preciso

#### **Lematizaci√≥n**
Reducir palabras a su forma can√≥nica (lema) usando an√°lisis ling√º√≠stico
- Ejemplo: `"corriendo"` ‚Üí `"correr"`
- M√°s lento pero m√°s preciso

#### **N-gramas**
Secuencias consecutivas de N palabras
- 1-grama (unigrama): `["el", "gato", "negro"]`
- 2-grama (bigrama): `["el gato", "gato negro"]`
- 3-grama (trigrama): `["el gato negro"]`

### ¬øPor qu√© fueron importantes?

- Reducen la dimensionalidad del vocabulario
- Normalizan variaciones morfol√≥gicas
- Permiten enfocarse en palabras significativas
- Base para modelos estad√≠sticos (Naive Bayes, clasificadores)

````python
# Importa la librer√≠a spaCy para procesamiento avanzado de lenguaje natural.
import spacy
# Importa la librer√≠a NLTK (Natural Language Toolkit) para tareas b√°sicas de NLP.
import nltk
# Importa la lista de "stopwords" (palabras comunes sin mucho significado) del m√≥dulo corpus de NLTK.
from nltk.corpus import stopwords
# Importa SnowballStemmer de NLTK, un algoritmo para reducir palabras a su ra√≠z (stemming).
from nltk.stem import SnowballStemmer
# Importa Counter del m√≥dulo collections, una clase para contar la frecuencia de elementos en una lista.
from collections import Counter
# Importa matplotlib.pyplot para la creaci√≥n de gr√°ficos y visualizaciones de datos.
import matplotlib.pyplot as plt

# Descargamos los recursos de NLTK, espec√≠ficamente la lista de stopwords para varios idiomas.
# `quiet=True` evita que se muestren mensajes de descarga en la consola.
nltk.download('stopwords', quiet=True)

# Cargamos el modelo de espa√±ol pre-entrenado de spaCy.
# `es_core_news_sm` es un modelo peque√±o que incluye componentes para tokenizaci√≥n, lematizaci√≥n,
# an√°lisis sint√°ctico y reconocimiento de entidades nombradas en espa√±ol.
# Se usa para procesar y analizar texto en espa√±ol de manera eficiente.
nlp = spacy.load('es_core_news_sm')

# Imprime un mensaje de confirmaci√≥n una vez que todas las librer√≠as y modelos han sido cargados.
print("‚úÖ Librer√≠as cargadas correctamente")
````

````python
# Define una cadena de texto multilinea que servir√° como ejemplo para las operaciones de NLP.
# Este texto se utilizar√° para demostrar tokenizaci√≥n, eliminaci√≥n de stopwords, lematizaci√≥n,
# y an√°lisis de frecuencia, entre otras t√©cnicas.
texto = """
El procesamiento del lenguaje natural es una disciplina fascinante que combina
la ling√º√≠stica computacional con la inteligencia artificial. Los investigadores
est√°n desarrollando sistemas cada vez m√°s sofisticados para analizar, comprender
y generar lenguaje humano. Estos sistemas procesan millones de textos y aprenden
patrones ling√º√≠sticos complejos.
"""

# Imprime un encabezado y el texto original, usando `.strip()` para eliminar espacios en blanco
# al inicio y al final del texto.
print("üìù Texto original:")
print(texto.strip())
````

### Tokenizaci√≥n con spaCy

````python
# Procesa el texto de ejemplo utilizando el objeto 'nlp' de spaCy cargado previamente.
# Esto convierte el texto en un objeto 'Doc' de spaCy, que contiene informaci√≥n
# detallada sobre cada token (palabra, puntuaci√≥n, espacio en blanco).
doc = nlp(texto)

# Extrae los tokens (unidades individuales de texto como palabras y signos de puntuaci√≥n)
# del objeto 'doc'. Se utiliza una comprensi√≥n de lista para obtener el texto de cada token.
tokens = [token.text for token in doc]

# Imprime un encabezado y la lista de tokens extra√≠dos.
print("üî§ Tokens extra√≠dos:")
print(tokens)
# Muestra el n√∫mero total de tokens encontrados en el texto.
print(f"\nüìä Total de tokens: {len(tokens)}")
````

### Eliminaci√≥n de stopwords

````python
# Carga el conjunto de stopwords (palabras muy comunes sin mucho significado, como "el", "la", "de")
# para el idioma espa√±ol desde NLTK. Se convierte a un `set` para b√∫squedas m√°s eficientes.
stop_words_es = set(stopwords.words('spanish'))

# Filtra los tokens obtenidos del texto original.
# Se incluyen solo aquellos tokens que:
# 1. No son stopwords (se comparan en min√∫sculas para asegurar consistencia).
# 2. Son alfab√©ticos (token.is_alpha), excluyendo as√≠ signos de puntuaci√≥n, n√∫meros, etc.
# Cada token se convierte a min√∫sculas antes de ser a√±adido a la lista.
tokens_filtrados = [
    token.text.lower()
    for token in doc
    if token.text.lower() not in stop_words_es
    and token.is_alpha  # Solo palabras (no puntuaci√≥n)
]

# Imprime los tokens resultantes despu√©s de la eliminaci√≥n de stopwords y puntuaci√≥n.
print("üßπ Tokens despu√©s de eliminar stopwords:")
print(tokens_filtrados)
# Muestra una comparaci√≥n entre la cantidad original de tokens y la cantidad despu√©s del filtrado,
# as√≠ como el porcentaje de reducci√≥n.
print(f"\nüìä Tokens originales: {len(tokens)} ‚Üí Tokens filtrados: {len(tokens_filtrados)}")
print(f"Reducci√≥n: {(1 - len(tokens_filtrados)/len(tokens)) * 100:.1f}%")
````

### Lematizaci√≥n con spaCy

````python
# Realiza la lematizaci√≥n, que es el proceso de reducir una palabra a su forma base o "lema"
# (por ejemplo, "corriendo" -> "correr", "casas" -> "casa").
# Filtra los tokens para incluir solo palabras alfab√©ticas y que no sean stopwords.
# `token.lemma_` es el atributo de spaCy que proporciona el lema de un token.
lemas = [token.lemma_ for token in doc if token.is_alpha and token.text.lower() not in stop_words_es]

# Imprime un encabezado para la comparaci√≥n.
print("üìö Comparaci√≥n Token ‚Üí Lema:\n")
# Crea una lista de tuplas que contienen el token original y su lema para los primeros 15 tokens alfab√©ticos.
# Esto se hace para visualizar el efecto de la lematizaci√≥n.
comparacion = [(token.text, token.lemma_) for token in doc if token.is_alpha][:15]

# Itera a trav√©s de la lista de comparaci√≥n para mostrar cada token y su lema.
for token, lema in comparacion:
    # Compara el token original (en min√∫sculas) con su lema (en min√∫sculas).
    if token.lower() != lema.lower():
        # Si son diferentes, significa que la lematizaci√≥n aplic√≥ un cambio.
        print(f"  {token:20} ‚Üí {lema}")
    else:
        # Si son iguales, el lema es el mismo que el token original.
        print(f"  {token:20} (sin cambio)")
````

### An√°lisis de frecuencias

````python
# Utiliza `collections.Counter` para contar la frecuencia de cada lema en la lista 'lemas'.
# Esto crea un diccionario donde las claves son los lemas y los valores son sus recuentos.
frecuencias = Counter(lemas)
# Obtiene los 10 lemas m√°s comunes y sus frecuencias.
# `most_common(10)` devuelve una lista de tuplas (lema, frecuencia).
mas_comunes = frecuencias.most_common(10)

# Imprime un encabezado para la lista de palabras m√°s frecuentes.
print("üìä Top 10 palabras m√°s frecuentes (lematizadas):\n")
# Itera sobre las palabras m√°s comunes para mostrarlas junto con su frecuencia.
# Se usa un formato para alinear las columnas y una representaci√≥n gr√°fica simple
# con bloques '‚ñà' para visualizar la frecuencia.
for palabra, freq in mas_comunes:
    print(f"  {palabra:20} {freq:3} veces {'‚ñà' * freq}")
````

### Generaci√≥n de N-gramas

````python
# Define una funci√≥n para generar n-gramas a partir de una lista de tokens.
# Un n-grama es una secuencia contigua de 'n' elementos de una muestra de texto.
def generar_ngramas(tokens, n):
    """Genera n-gramas a partir de una lista de tokens"""
    ngramas = [] # Inicializa una lista vac√≠a para almacenar los n-gramas.
    # Itera sobre la lista de tokens, deteni√©ndose en el punto donde ya no es posible formar un n-grama completo.
    for i in range(len(tokens) - n + 1):
        # Toma una sub-secuencia de 'n' tokens comenzando desde la posici√≥n 'i'.
        # Une estos tokens con un espacio para formar un solo string que representa el n-grama.
        ngrama = ' '.join(tokens[i:i+n])
        ngramas.append(ngrama) # A√±ade el n-grama formado a la lista.
    return ngramas # Devuelve la lista de todos los n-gramas generados.

# Generamos bigramas (n=2) y trigramas (n=3) utilizando la funci√≥n `generar_ngramas`.
# Se utilizan los `tokens_filtrados` (sin stopwords ni puntuaci√≥n) para obtener n-gramas m√°s significativos.
bigramas = generar_ngramas(tokens_filtrados, 2) # Secuencias de 2 palabras.
trigramas = generar_ngramas(tokens_filtrados, 3) # Secuencias de 3 palabras.

# Imprime un encabezado y los primeros 10 bigramas para visualizaci√≥n.
print("üìå Ejemplos de Bigramas (2-gramas):\n")
for bg in bigramas[:10]:
    print(f"  ‚Ä¢ {bg}")

# Imprime un encabezado y los primeros 8 trigramas para visualizaci√≥n.
print("\nüìå Ejemplos de Trigramas (3-gramas):\n")
for tg in trigramas[:8]:
    print(f"  ‚Ä¢ {tg}")
````

### üìù EJERCICIO 2: Stemming vs Lematizaci√≥n

````python
# üìù EJERCICIO: Implementa stemming y comp√°ralo con lematizaci√≥n

# Define una lista de palabras de prueba para demostrar y comparar el stemming y la lematizaci√≥n.
palabras_prueba = [
    "corriendo", "corri√≥", "correr", "corredor",
    "desarrollando", "desarrolladores", "desarrollo",
    "sistemas", "sistem√°tico", "sistematizaci√≥n"
]

# Inicializa un objeto SnowballStemmer para el idioma espa√±ol.
# El SnowballStemmer es un algoritmo de stemming que reduce las palabras a su ra√≠z
# eliminando sufijos, a menudo resultando en "ra√≠ces" que no son palabras v√°lidas por s√≠ mismas.
stemmer = SnowballStemmer('spanish')

# Imprime un encabezado para la tabla comparativa.
print("üîç Comparaci√≥n Stemming vs Lematizaci√≥n:\n")
# Imprime los encabezados de las columnas para una mejor legibilidad.
print(f"{'Palabra Original':<25} {'Stemming':<20} {'Lematizaci√≥n':<20}")
print("-" * 70) # L√≠nea divisoria.

# Itera sobre cada palabra en la lista de prueba.
for palabra in palabras_prueba:
    # Aplica stemming a la palabra actual utilizando el stemmer inicializado.
    stem = stemmer.stem(palabra)

    # Aplica lematizaci√≥n a la palabra.
    # Primero, procesa la palabra con el modelo spaCy para obtener un objeto 'Doc'.
    doc_temp = nlp(palabra)
    # Luego, extrae el lema del primer (y √∫nico) token en el 'Doc'.
    # La lematizaci√≥n busca la forma base real de una palabra (lema v√°lido en el idioma).
    lema = doc_temp[0].lemma_

    # Imprime la palabra original, su versi√≥n stemmed y su lema, formateados para alineaci√≥n.
    print(f"{'palabra':<25} {stem:<20} {lema:<20}")

# Imprime observaciones finales sobre las diferencias entre stemming y lematizaci√≥n.
print("\nüí° Observa las diferencias:")
print("   - Stemming: m√°s agresivo, puede producir ra√≠ces no v√°lidas")
print("   - Lematizaci√≥n: m√°s preciso, produce palabras reales")
````

### üìù EJERCICIO 3: An√°lisis de N-gramas

````python
# üìù EJERCICIO: Analiza las diferencias entre unigramas, bigramas y trigramas

# Generamos los tres tipos de n-gramas a partir de los tokens filtrados.
# Unigramas: Son simplemente las palabras individuales (tokens filtrados).
unigramas = tokens_filtrados
# Bigramas: Secuencias de dos palabras consecutivas, generadas con la funci√≥n `generar_ngramas`.
bigramas = generar_ngramas(tokens_filtrados, 2)
# Trigramas: Secuencias de tres palabras consecutivas.
trigramas = generar_ngramas(tokens_filtrados, 3)

# Contamos las frecuencias de cada tipo de n-grama utilizando `collections.Counter`.
# Esto nos ayuda a identificar los n-gramas m√°s comunes en cada categor√≠a.
freq_uni = Counter(unigramas) # Frecuencias para unigramas.
freq_bi = Counter(bigramas)   # Frecuencias para bigramas.
freq_tri = Counter(trigramas) # Frecuencias para trigramas.

# Imprime un encabezado para el an√°lisis comparativo.
print("üìä An√°lisis comparativo de N-gramas\n")
print("="*80)

# Muestra los 5 unigramas m√°s frecuentes.
print("\nüîπ Top 5 Unigramas (palabras individuales):")
for ngrama, freq in freq_uni.most_common(5):
    print(f"  {freq}x - {ngrama}")

# Muestra los 5 bigramas m√°s frecuentes.
print("\nüîπ Top 5 Bigramas (pares de palabras):")
for ngrama, freq in freq_bi.most_common(5):
    print(f"  {freq}x - {ngrama}")

# Muestra los 5 trigramas m√°s frecuentes.
print("\nüîπ Top 5 Trigramas (tr√≠os de palabras):")
for ngrama, freq in freq_tri.most_common(5):
    print(f"  {freq}x - {ngrama}")

print("\n" + "="*80)
# Ofrece algunas observaciones clave sobre las caracter√≠sticas de cada tipo de n-grama.
print("\nüí° ¬øQu√© observas?")
print("   - Unigramas: palabras aisladas, pierden contexto")
print("   - Bigramas: capturan relaciones b√°sicas entre palabras")
print("   - Trigramas: m√°s contexto, pero tambi√©n m√°s escasos")
````

## 3. Representaciones Vectoriales: Bag of Words y TF-IDF

### ¬øPor qu√© vectorizar el texto?

Los algoritmos de machine learning **no entienden texto**, solo n√∫meros. Necesitamos convertir el texto en vectores num√©ricos.

---

### **Bag of Words (BoW) - Bolsa de Palabras**

**Concepto**: Representar un documento como un vector de frecuencias de palabras, ignorando el orden.

**Ejemplo**:
```
Documento 1: "el gato come pescado"
Documento 2: "el perro come carne"

Vocabulario: [el, gato, perro, come, pescado, carne]

Doc1: [1, 1, 0, 1, 1, 0]
Doc2: [1, 0, 1, 1, 0, 1]
```

**Ventajas**:
- Simple de implementar
- Funciona bien para tareas b√°sicas

**Limitaciones**:
- Pierde el orden de las palabras
- No captura significado sem√°ntico
- Palabras comunes dominan la representaci√≥n

---

### **TF-IDF (Term Frequency - Inverse Document Frequency)**

**Concepto**: Ponderar las palabras seg√∫n su importancia en el documento y en todo el corpus.

**F√≥rmula**:
```
TF-IDF(palabra, doc) = TF(palabra, doc) √ó IDF(palabra)

TF = Frecuencia de la palabra en el documento
IDF = log(Total documentos / Documentos que contienen la palabra)
```

**Intuici√≥n**:
- Palabras **frecuentes** en un documento pero **raras** en el corpus ‚Üí **Mayor peso**
- Palabras **muy comunes** en todos los documentos ‚Üí **Menor peso**

**Ventajas**:
- Reduce el peso de palabras comunes
- Resalta palabras discriminativas
- Mejor para b√∫squeda y clasificaci√≥n

**Limitaciones**:
- Sigue sin capturar sem√°ntica
- Sensible al tama√±o del corpus

---

````python
# Importa CountVectorizer y TfidfVectorizer de sklearn.feature_extraction.text.
# Estas clases se utilizan para convertir una colecci√≥n de documentos de texto
# en matrices de recuentos de tokens (CountVectorizer) o en matrices TF-IDF (TfidfVectorizer).
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
# Importa pandas para la manipulaci√≥n y an√°lisis de datos, especialmente para DataFrames.
import pandas as pd
# Importa numpy para operaciones num√©ricas, especialmente con arrays.
import numpy as np

# Define un corpus de ejemplo (una colecci√≥n de documentos) que se utilizar√°
# para demostrar Bag of Words y TF-IDF. Cada elemento de la lista es un documento.
corpus = [
    "El aprendizaje autom√°tico es una rama de la inteligencia artificial",
    "Las redes neuronales profundas son la base del deep learning",
    "Los modelos de lenguaje procesan y generan texto humano",
    "El procesamiento de lenguaje natural combina ling√º√≠stica e inteligencia artificial",
    "Los transformers revolucionaron el procesamiento del lenguaje natural"
]

# Imprime un encabezado y enumera cada documento en el corpus.
print("üìö Corpus de documentos:\n")
for i, doc in enumerate(corpus, 1):
    print(f"{i}. {doc}")
````

### Bag of Words (BoW)

````python
# Crea una instancia de CountVectorizer, que implementa la estrategia Bag of Words (BoW).
# Esta clase convierte una colecci√≥n de documentos de texto en una matriz de recuentos de tokens.
vectorizer_bow = CountVectorizer()

# Aplica el vectorizador al corpus.
# `fit_transform()` primero aprende el vocabulario del corpus (`fit`)
# y luego transforma los documentos en una matriz de recuentos de t√©rminos (`transform`).
bow_matrix = vectorizer_bow.fit_transform(corpus)

# Obtiene las palabras (features) que forman el vocabulario aprendido por el vectorizador.
# Estas ser√°n las columnas de la matriz BoW.
vocabulario = vectorizer_bow.get_feature_names_out()

# Convierte la matriz BoW (que es un formato disperso para ahorrar memoria) a un array denso,
# y luego a un DataFrame de pandas para una visualizaci√≥n m√°s legible.
df_bow = pd.DataFrame(
    bow_matrix.toarray(), # `toarray()` convierte la matriz dispersa a un array numpy.
    columns=vocabulario, # Asigna el vocabulario como nombres de columna.
    index=[f"Doc{i+1}" for i in range(len(corpus))] # Asigna nombres a las filas (documentos).
)

# Imprime un encabezado y la matriz Bag of Words resultante.
print("üéí Matriz Bag of Words:\n")
print(df_bow)
# Muestra el n√∫mero total de palabras √∫nicas (tama√±o del vocabulario) identificadas.
print(f"\nüìä Tama√±o del vocabulario: {len(vocabulario)} palabras")
# Muestra las dimensiones de la matriz BoW (n√∫mero de documentos x tama√±o del vocabulario).
print(f"üìä Dimensi√≥n de la matriz: {bow_matrix.shape}")
````

### TF-IDF

````python
# Crea una instancia de TfidfVectorizer.
# Este vectorizador calcula la frecuencia de t√©rmino (TF) y la frecuencia inversa
# de documento (IDF) para cada palabra en el corpus, asignando un peso a cada palabra
# que refleja su importancia en un documento y en el corpus general.
vectorizer_tfidf = TfidfVectorizer()

# Aplica el vectorizador TF-IDF al corpus.
# `fit_transform()` aprende el vocabulario y las estad√≠sticas IDF del corpus,
# luego transforma los documentos en una matriz TF-IDF.
tfidf_matrix = vectorizer_tfidf.fit_transform(corpus)

# Convierte la matriz TF-IDF (dispersa) a un array denso y luego a un DataFrame de pandas.
df_tfidf = pd.DataFrame(
    tfidf_matrix.toarray(), # Convierte la matriz dispersa a un array numpy.
    columns=vectorizer_tfidf.get_feature_names_out(), # Obtiene las palabras del vocabulario como columnas.
    index=[f"Doc{i+1}" for i in range(len(corpus))] # Asigna nombres a las filas (documentos).
)

# Imprime un encabezado y la matriz TF-IDF resultante, redondeando los valores a 3 decimales
# para una mejor legibilidad.
print("‚öñÔ∏è Matriz TF-IDF:\n")
print(df_tfidf.round(3))
````

### An√°lisis de palabras m√°s relevantes por documento

````python
# Itera a trav√©s de cada documento en el DataFrame TF-IDF para identificar las palabras m√°s relevantes.
print("üîù Top 5 palabras m√°s relevantes por documento (TF-IDF):\n")

# `enumerate(df_tfidf.index)` permite iterar tanto por el √≠ndice (nombre del documento)
# como por su posici√≥n (i) en la lista de √≠ndices.
for i, doc_name in enumerate(df_tfidf.index):
    # Selecciona la fila correspondiente al documento actual (`doc_name`),
    # ordena sus valores de TF-IDF de mayor a menor (`sort_values(ascending=False)`),
    # y toma las 5 primeras palabras (`head(5)`).
    top_palabras = df_tfidf.loc[doc_name].sort_values(ascending=False).head(5)

    # Imprime el nombre del documento y una parte de su contenido original para referencia.
    print(f"üìÑ {doc_name}: {corpus[i][:60]}...")
    print("   Palabras clave:")
    # Itera sobre las palabras con los mayores scores TF-IDF para ese documento.
    for palabra, score in top_palabras.items():
        # Asegura que solo se impriman palabras con un score TF-IDF positivo (es decir, que aparecen en el documento).
        if score > 0:
            # Imprime la palabra y su score TF-IDF, formateado para una mejor presentaci√≥n.
            print(f"      ‚Ä¢ {palabra:20} (score: {score:.3f})")
    print() # Imprime una l√≠nea en blanco para separar los resultados de cada documento.
````

## 4. Embeddings Modernos

### Del espacio simb√≥lico al espacio sem√°ntico

Las representaciones anteriores (BoW, TF-IDF) tienen una **limitaci√≥n fundamental**:
- Tratan las palabras como s√≠mbolos independientes
- No capturan relaciones sem√°nticas
- "rey" y "monarca" son tan diferentes como "rey" y "manzana"

---

### ¬øQu√© son los embeddings?

**Embeddings** = Representaciones vectoriales **densas** que capturan el **significado sem√°ntico**

**Caracter√≠sticas**:
- Vectores de dimensi√≥n fija (ej: 300, 768, 1024)
- Palabras similares ‚Üí vectores cercanos en el espacio
- Permiten aritm√©tica sem√°ntica: `Rey - Hombre + Mujer ‚âà Reina`

---

### Evoluci√≥n de los embeddings

#### **1. Word2Vec (2013)**
- **Idea**: "Una palabra se entiende por las palabras que la rodean"
- **Arquitecturas**: CBOW (Continuous Bag of Words), Skip-gram
- **Limitaci√≥n**: Un embedding fijo por palabra (no contextual)

#### **2. GloVe (2014)**
- Global Vectors for Word Representation
- Basado en matriz de co-ocurrencias globales
- Captura tanto estad√≠sticas locales como globales

#### **3. Embeddings Contextualizados (2018+)**
- **BERT, GPT, RoBERTa**: Embeddings que cambian seg√∫n el contexto
- Ejemplo: "banco" (instituci√≥n financiera) vs "banco" (asiento)
  - Antes: mismo vector para ambos
  - Ahora: vectores diferentes seg√∫n contexto

#### **4. Sentence Transformers (2019+)**
- Embeddings para **frases completas**, no solo palabras
- Optimizados para tareas de similitud sem√°ntica
- Base de sistemas de b√∫squeda sem√°ntica y RAG

---

### Aplicaciones

- **B√∫squeda sem√°ntica**: Encontrar documentos similares
- **Sistemas de recomendaci√≥n**: Productos, contenidos
- **Clustering**: Agrupar textos por tema
- **Clasificaci√≥n**: Input para modelos de ML
- **RAG**: Recuperaci√≥n de contexto relevante

````python
# Importa `SentenceTransformer` para cargar modelos de embeddings de frases.
from sentence_transformers import SentenceTransformer
# Importa `cosine_similarity` para calcular la similitud entre vectores.
# Se usa para determinar qu√© tan similares son sem√°nticamente dos frases.
from sklearn.metrics.pairwise import cosine_similarity
# Importa PCA (Principal Component Analysis) para reducci√≥n de dimensionalidad.
# Se utilizar√° para visualizar embeddings de alta dimensi√≥n en 2D.
from sklearn.decomposition import PCA
# Importa numpy para operaciones num√©ricas y manipulaci√≥n de arrays.
import numpy as np

# Carga un modelo pre-entrenado de SentenceTransformer.
# 'paraphrase-multilingual-MiniLM-L12-v2' es un modelo multiling√ºe eficiente
# que genera embeddings de frases, ideal para tareas de similitud sem√°ntica.
# Est√° optimizado para un buen rendimiento en varios idiomas, incluido el espa√±ol.
modelo_embeddings = SentenceTransformer('paraphrase-multilingual-MiniLM-L12-v2')

# Imprime un mensaje de confirmaci√≥n de que el modelo ha sido cargado.
print("‚úÖ Modelo de embeddings cargado")
# Muestra la dimensi√≥n de los vectores de embeddings que produce este modelo.
# Cada frase se convertir√° en un vector de esta longitud.
print(f"üìä Dimensi√≥n de los embeddings: {modelo_embeddings.get_sentence_embedding_dimension()}")
````

````python
# Define una lista de frases de ejemplo que se utilizar√°n para generar embeddings.
# Estas frases cubren diferentes temas para demostrar c√≥mo los embeddings capturan
# la similitud sem√°ntica.
frases = [
    "El perro corre por el parque",
    "Un canino est√° corriendo en el jard√≠n",
    "El gato duerme en el sof√°",
    "Python es un lenguaje de programaci√≥n",
    "Machine learning es una rama de la inteligencia artificial",
    "La IA est√° transformando el mundo",
    "Me gusta comer pizza",
    "La pizza es mi comida favorita"
]

# Imprime un encabezado y enumera cada frase de entrada para que el usuario pueda verlas.
print("üìù Frases de entrada:\n")
for i, frase in enumerate(frases, 1):
    print(f"{i}. {frase}")
````

````python
# Genera los embeddings para cada frase en la lista `frases` utilizando el modelo cargado.
# `modelo_embeddings.encode()` convierte cada frase en un vector num√©rico de alta dimensi√≥n.
embeddings = modelo_embeddings.encode(frases)

# Imprime un mensaje de confirmaci√≥n de que los embeddings han sido generados.
print(f"‚úÖ Embeddings generados")
# Muestra la forma de la matriz de embeddings, indicando el n√∫mero de frases
# y la dimensi√≥n de cada vector de embedding.
print(f"üìä Forma de la matriz: {embeddings.shape}")
print(f"    {len(frases)} frases √ó {embeddings.shape[1]} dimensiones")

# Muestra un ejemplo del embedding de la primera frase, incluyendo sus primeros 10 valores.
# Esto ayuda a entender la naturaleza num√©rica de los embeddings.
print(f"\nüîç Ejemplo - Embedding de la primera frase:")
print(f"   '{frases[0]}' ")
print(f"   Primeros 10 valores: {embeddings[0][:10]}")
````

### C√°lculo de similitud sem√°ntica

````python
# Calcula la similitud del coseno entre todos los pares de embeddings generados.
# La similitud del coseno mide el √°ngulo entre dos vectores; un valor cercano a 1 indica
# alta similitud sem√°ntica, mientras que un valor cercano a 0 (o negativo) indica baja similitud.
matriz_similitud = cosine_similarity(embeddings)

# Crea un DataFrame de pandas a partir de la matriz de similitud.
# Esto mejora la visualizaci√≥n, permitiendo ver las similitudes de manera tabular.
df_similitud = pd.DataFrame(
    matriz_similitud,
    index=[f"F{i+1}" for i in range(len(frases))], # Asigna etiquetas de fila como "F1", "F2", etc.
    columns=[f"F{i+1}" for i in range(len(frases))] # Asigna etiquetas de columna de manera similar.
)

# Imprime la matriz de similitud sem√°ntica, redondeando los valores a 3 decimales para claridad.
# Se incluye una nota explicativa sobre la interpretaci√≥n de los valores.
print("üìä Matriz de Similitud Sem√°ntica (0=diferentes, 1=id√©nticas):\n")
print(df_similitud.round(3))
````

````python
# Visualiza la matriz de similitud del coseno utilizando un mapa de calor (heatmap).
# Esto proporciona una representaci√≥n gr√°fica intuitiva de las similitudes entre frases.
plt.figure(figsize=(10, 8)) # Crea una figura con un tama√±o espec√≠fico.
# `plt.imshow()` muestra la matriz como una imagen.
# `cmap='RdYlGn'` utiliza un mapa de colores que va de rojo (baja similitud) a verde (alta similitud),
# pasando por amarillo.
# `interpolation='nearest'` para una visualizaci√≥n de celdas claras.
# `vmin=0, vmax=1` establece los l√≠mites de color para corresponder con los valores de similitud del coseno.
plt.imshow(matriz_similitud, cmap='RdYlGn', interpolation='nearest', vmin=0, vmax=1)
# A√±ade una barra de color lateral que explica la escala de similitud.
plt.colorbar(label='Similitud del Coseno', shrink=0.8)
# Establece el t√≠tulo del mapa de calor.
plt.title('Mapa de Calor - Similitud Sem√°ntica entre Frases', fontsize=14, fontweight='bold', pad=20)
# Configura las etiquetas del eje X con los nombres de las frases (F1, F2, etc.).
plt.xticks(range(len(frases)), [f"F{i+1}" for i in range(len(frases))])
# Configura las etiquetas del eje Y con los nombres de las frases.
plt.yticks(range(len(frases)), [f"F{i+1}" for i in range(len(frases))])

# A√±ade los valores num√©ricos de similitud directamente en cada celda del mapa de calor.
for i in range(len(frases)):
    for j in range(len(frases)):
        valor = matriz_similitud[i, j]
        # Elige el color del texto (blanco o negro) para asegurar buena visibilidad
        # dependiendo del color de fondo de la celda.
        color = 'white' if valor > 0.7 else 'black'
        # `plt.text()` coloca el valor formateado en la posici√≥n (j, i) de la celda.
        plt.text(j, i, f'{valor:.2f}', ha='center', va='center', color=color, fontsize=9)

# Ajusta el dise√±o de la figura para evitar que los elementos se superpongan.
plt.tight_layout()
# Muestra el mapa de calor.
plt.show()

# Imprime una leyenda para que el usuario pueda identificar qu√© frase corresponde a cada "F" en el gr√°fico.
print("\nüìù Leyenda de frases:")
for i, frase in enumerate(frases, 1):
    print(f"F{i}: {frase}")
````

````python
# Este bloque de c√≥digo busca e imprime los 5 pares de frases con mayor similitud sem√°ntica.
print("üîù Top 5 pares de frases m√°s similares:\n")

# Inicializa una lista para almacenar tuplas de (√≠ndice_frase1, √≠ndice_frase2, valor_similitud).
pares_similitud = []
# Itera sobre todas las frases para formar pares √∫nicos.
for i in range(len(frases)):
    # Itera desde `i+1` para evitar duplicados (ej. (F1, F2) y (F2, F1))
    # y para excluir la similitud de una frase consigo misma (que siempre es 1).
    for j in range(i+1, len(frases)):
        # A√±ade la tupla (√≠ndice de la primera frase, √≠ndice de la segunda frase, similitud) a la lista.
        pares_similitud.append((i, j, matriz_similitud[i, j]))

# Ordena la lista `pares_similitud` en orden descendente bas√°ndose en el valor de similitud.
# `key=lambda x: x[2]` especifica que la ordenaci√≥n se realice por el tercer elemento de cada tupla (el score de similitud).
pares_similitud.sort(key=lambda x: x[2], reverse=True)

# Muestra los 5 pares de frases con mayor similitud.
for idx, (i, j, sim) in enumerate(pares_similitud[:5], 1):
    print(f"{idx}. Similitud: {sim:.4f}") # Imprime el rango y el valor de similitud.
    print(f"   ‚Ä¢ {frases[i]}") # Imprime la primera frase del par.
    print(f"   ‚Ä¢ {frases[j]}") # Imprime la segunda frase del par.
    print() # A√±ade una l√≠nea en blanco para separar los resultados de cada par.
````

### üìù EJERCICIO 9: B√∫squeda sem√°ntica

````python
# üìù EJERCICIO: Crea una funci√≥n de b√∫squeda sem√°ntica

# Define una funci√≥n `buscar_frase_similar` que toma una consulta (query)
# y busca las frases m√°s sem√°nticamente similares dentro de una base de datos.
def buscar_frase_similar(query, frases_bd, embeddings_bd, top_k=3):
    """
    Encuentra las frases m√°s similares a una query.

    Args:
        query: Texto de b√∫squeda (la frase que queremos comparar).
        frases_bd: Lista de frases que componen la base de datos.
        embeddings_bd: Los embeddings precalculados de las frases en la base de datos.
        top_k: N√∫mero de resultados m√°s similares a retornar.

    Returns:
        Lista de tuplas (frase, similitud) de las frases m√°s relevantes.
    """
    # Genera el embedding de la consulta utilizando el `modelo_embeddings` global.
    # El input se envuelve en una lista `[query]` porque `encode` espera una lista de frases.
    query_embedding = modelo_embeddings.encode([query])

    # Calcula la similitud del coseno entre el embedding de la consulta y todos los embeddings
    # de la base de datos. `[0]` se usa para extraer el array de similitudes.
    similitudes = cosine_similarity(query_embedding, embeddings_bd)[0]

    # Ordena los √≠ndices de similitud de mayor a menor y selecciona los `top_k` √≠ndices.
    # `np.argsort(similitudes)` devuelve los √≠ndices que ordenar√≠an el array.
    # `[::-1]` invierte este orden para tener los m√°s altos primero.
    # `[:top_k]` selecciona los `top_k` primeros √≠ndices.
    indices_top = np.argsort(similitudes)[::-1][:top_k]

    # Construye la lista de resultados, que contiene tuplas de (frase, similitud)
    # para los `top_k` resultados.
    resultados = [(frases_bd[idx], similitudes[idx]) for idx in indices_top]

    return resultados # Devuelve la lista de resultados.

# Define una consulta de prueba para la funci√≥n de b√∫squeda.
consulta = "animales dom√©sticos"  # Cambia esta consulta para probar otras b√∫squedas.

# Imprime la consulta y un encabezado para los resultados.
print(f"üîé B√∫squeda: '{consulta}'\n")
print("Resultados m√°s relevantes:\n")

# Llama a la funci√≥n de b√∫squeda con la consulta, las frases y embeddings existentes,
# y pide los 5 resultados m√°s relevantes (`top_k=5`).
resultados = buscar_frase_similar(consulta, frases, embeddings, top_k=5)

# Imprime cada resultado, incluyendo su n√∫mero, score de similitud y la frase.
for i, (frase, sim) in enumerate(resultados, 1):
    print(f"{i}. [{sim:.4f}] {frase}")

print("\nüí° Prueba con otras consultas cambiando la variable 'consulta'")
````

### üìù EJERCICIO 10: Clustering de frases por tema

````python
# üìù EJERCICIO: Agrupa las frases por tema usando K-means

# Importa KMeans del m√≥dulo cluster de scikit-learn.
# KMeans es un algoritmo de clustering popular que agrupa puntos de datos en 'k' clusters
# bas√°ndose en su similitud. En este caso, agrupar√° frases por su similitud sem√°ntica
# a trav√©s de sus embeddings.
from sklearn.cluster import KMeans

# Define el n√∫mero deseado de clusters (grupos de temas).
n_clusters = 3  # Puedes cambiar este n√∫mero para experimentar con m√°s o menos grupos.

# Aplica el algoritmo K-means a los embeddings de las frases.
# `KMeans(n_clusters=n_clusters, random_state=42)` inicializa el modelo.
# `random_state` asegura la reproducibilidad de los resultados.
# `fit_predict(embeddings)` ajusta el modelo a los embeddings y asigna cada embedding
# a un cluster, devolviendo un array con la ID del cluster para cada frase.
kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init='auto') # A√±adido n_init='auto' para suprimir warning
clusters = kmeans.fit_predict(embeddings)

print(f"üìä Agrupaci√≥n de frases en {n_clusters} clusters (temas):\n")
print("="*80)

# Itera sobre cada ID de cluster para mostrar las frases asignadas a cada grupo.
for cluster_id in range(n_clusters):
    print(f"\nüè∑Ô∏è Cluster {cluster_id + 1}:")
    # Filtra los √≠ndices de las frases que pertenecen al cluster actual.
    indices = [i for i, c in enumerate(clusters) if c == cluster_id]
    # Imprime cada frase del cluster.
    for idx in indices:
        print(f"   ‚Ä¢ {frases[idx]}")

# Visualiza los embeddings en 2D (previamente reducidos con PCA) coloreados seg√∫n su cluster.
plt.figure(figsize=(12, 8)) # Crea una figura para la visualizaci√≥n.
# `plt.cm.Set3(clusters)` asigna un color diferente del mapa de colores 'Set3' a cada cluster ID.
colores = plt.cm.Set3(clusters)

# Dibuja un gr√°fico de dispersi√≥n de los embeddings 2D, donde el color de cada punto
# indica a qu√© cluster pertenece la frase.
plt.scatter(embeddings_2d[:, 0], embeddings_2d[:, 1], s=300, c=colores, edgecolors='black', linewidths=2)

# A√±ade etiquetas a cada punto en el gr√°fico, mostrando la etiqueta de la frase (F#)
# y el cluster al que pertenece (C#).
for i, frase in enumerate(frases):
    plt.annotate(
        f"F{i+1}\n(C{clusters[i]+1})", # Formato de la etiqueta.
        (embeddings_2d[i, 0], embeddings_2d[i, 1]), # Posici√≥n de la etiqueta.
        fontsize=9, ha='center', va='center' # Estilo.
    )

# Establece las etiquetas de los ejes y el t√≠tulo del gr√°fico de clustering.
plt.xlabel('Componente Principal 1', fontsize=12)
plt.ylabel('Componente Principal 2', fontsize=12)
plt.title(f'Clustering de Frases por Similitud Sem√°ntica ({n_clusters} clusters)', fontsize=14, fontweight='bold')
# A√±ade una cuadr√≠cula suave.
plt.grid(alpha=0.3)
# Ajusta el dise√±o.
plt.tight_layout()
# Muestra el gr√°fico.
plt.show()
````

## 5. Clasificaci√≥n con Modelos Tipo BERT (Encoder)

### Arquitecturas Transformer: Encoder vs Decoder

Los transformers se pueden clasificar en tres categor√≠as seg√∫n su arquitectura:

---

### **üîç Encoder-only (BERT, RoBERTa, DistilBERT)**

**Funci√≥n**: **Comprensi√≥n** y **representaci√≥n** de texto

**Caracter√≠sticas**:
- Atenci√≥n **bidireccional**: Lee el texto completo en ambas direcciones
- Genera representaciones contextualizadas de cada token
- No dise√±ado para generar texto nuevo

**Usos principales**:
- ‚úÖ Clasificaci√≥n de texto (sentimientos, spam, categor√≠as)
- ‚úÖ Named Entity Recognition (NER)
- ‚úÖ Question Answering
- ‚úÖ Extracci√≥n de informaci√≥n
- ‚úÖ An√°lisis de similitud

**Ejemplo**: BERT
- Pre-entrenado con Masked Language Modeling (MLM)
- Predice palabras ocultas en un contexto bidireccional

---

### **‚úçÔ∏è Decoder-only (GPT, LLaMA, Claude)**

**Funci√≥n**: **Generaci√≥n** de texto

**Caracter√≠sticas**:
- Atenci√≥n **unidireccional** (solo hacia la izquierda)
- Predice el siguiente token bas√°ndose en los anteriores
- Autore Gresivo: genera palabra por palabra

**Usos principales**:
- ‚úÖ Generaci√≥n de texto (historias, art√≠culos)
- ‚úÖ Chatbots y asistentes conversacionales
- ‚úÖ Completado de c√≥digo
- ‚úÖ Traducci√≥n (con prompt adecuado)
- ‚úÖ Res√∫menes

**Ejemplo**: GPT-3/4
- Pre-entrenado con Causal Language Modeling
- Predice la siguiente palabra dadas las anteriores

---

### **üîÑ Encoder-Decoder (T5, BART, mBART)**

**Funci√≥n**: **Transformaci√≥n** de texto (input ‚Üí output)

**Caracter√≠sticas**:
- Encoder: comprende el input
- Decoder: genera el output
- Ideal cuando input y output son diferentes

**Usos principales**:
- ‚úÖ Traducci√≥n autom√°tica
- ‚úÖ Res√∫menes (texto largo ‚Üí resumen corto)
- ‚úÖ Par√°frasis
- ‚úÖ Correcci√≥n gramatical

---

### Tabla Comparativa

| Arquitectura | Atenci√≥n | Tarea Principal | Ejemplos | Mejor para |
|--------------|----------|-----------------|----------|------------|
| Encoder | Bidireccional | Comprensi√≥n | BERT, RoBERTa | Clasificaci√≥n, NER |
| Decoder | Unidireccional | Generaci√≥n | GPT, LLaMA | Chatbots, escritura |
| Encoder-Decoder | Ambas | Transformaci√≥n | T5, BART | Traducci√≥n, resumen |

````python
# Importa la funci√≥n `pipeline` de la librer√≠a `transformers`.
# `pipeline` es una forma sencilla de usar modelos pre-entrenados para tareas espec√≠ficas
# sin tener que preocuparse por la tokenizaci√≥n o el post-procesamiento.
from transformers import pipeline

# Carga un modelo pre-entrenado para la tarea de "sentiment-analysis" (an√°lisis de sentimientos).
# El modelo especificado, "nlptown/bert-base-multilingual-uncased-sentiment", es un modelo BERT
# multiling√ºe que ha sido ajustado (fine-tuneado) para predecir el sentimiento en una escala de 1 a 5 estrellas,
# donde 1 es muy negativo y 5 es muy positivo.
# `device=-1` especifica que el modelo se ejecutar√° en la CPU. Si hay una GPU disponible y se desea usar,
# se puede cambiar a `device=0`.
clasificador_sentimientos = pipeline(
    "sentiment-analysis",
    model="nlptown/bert-base-multilingual-uncased-sentiment",
    device=-1  # CPU (-1), cambiar a 0 para GPU (e.g., if a CUDA-enabled GPU is available)
)

# Imprime un mensaje de confirmaci√≥n cuando el modelo ha sido cargado exitosamente.
print("‚úÖ Modelo de clasificaci√≥n de sentimientos cargado")
````

````python
# Define una lista de frases de ejemplo que ser√°n utilizadas para el an√°lisis de sentimientos.
# Estas frases cubren un rango de sentimientos desde muy positivos hasta muy negativos,
# as√≠ como algunos neutrales o mixtos, para demostrar la capacidad del clasificador.
frases_sentimientos = [
    "Esta pel√≠cula es absolutamente incre√≠ble, la mejor que he visto en a√±os",
    "Me encant√≥ el servicio, muy profesional y r√°pido",
    "El producto es aceptable, cumple con lo esperado",
    "No me gust√≥ para nada, muy decepcionante",
    "P√©sima experiencia, nunca volver√© a comprar aqu√≠",
    "El hotel est√° bien ubicado pero las habitaciones son peque√±as",
    "Excelente calidad-precio, totalmente recomendable",
    "El libro es interesante aunque un poco largo",
    "Horrible, una p√©rdida total de tiempo y dinero",
    "Fant√°stico, super√≥ todas mis expectativas"
]

# Imprime un encabezado y enumera cada frase que se va a clasificar.
# `f"{i:2}. {frase}"` formatea el n√∫mero con dos espacios para alineaci√≥n.
print("üìù Frases a clasificar:\n")
for i, frase in enumerate(frases_sentimientos, 1):
    print(f"{i:2}. {frase}")
````

````python
# Clasifica todas las frases de la lista `frases_sentimientos` utilizando el `pipeline` de sentimiento.
# `clasificador_sentimientos` procesa las frases y devuelve una lista de diccionarios,
# donde cada diccionario contiene la etiqueta predicha (ej. '5 stars') y el score de confianza.
resultados = clasificador_sentimientos(frases_sentimientos)

print("üéØ Resultados de la Clasificaci√≥n:\n")
print("="*90) # L√≠nea divisoria para formato.

# Itera a trav√©s de las frases originales y sus resultados de clasificaci√≥n.
for i, (frase, resultado) in enumerate(zip(frases_sentimientos, resultados), 1):
    label = resultado['label']   # Extrae la etiqueta predicha (ej. '5 stars').
    score = resultado['score']   # Extrae el score de confianza de la predicci√≥n.
    # Convierte la parte num√©rica de la etiqueta (ej. '5' de '5 stars') a un entero.
    estrellas_num = int(label.split()[0])

    # Asigna una categor√≠a sem√°ntica (Negativo, Neutral, Positivo) y un emoji de color
    # bas√°ndose en el n√∫mero de estrellas predicho.
    if estrellas_num <= 2:
        categoria = "üòû Negativo"
        color = "üî¥"
    elif estrellas_num == 3:
        categoria = "üòê Neutral"
        color = "üü°"
    else:
        categoria = "üòä Positivo"
        color = "üü¢"

    # Imprime la frase, trunc√°ndola si es muy larga para mantener la legibilidad.
    print(f"\n{i:2}. {frase[:70]}..." if len(frase) > 70 else f"\n{i:2}. {frase}")
    # Imprime el resultado de la clasificaci√≥n: emoji, categor√≠a, etiqueta original,
    # score de confianza y una representaci√≥n visual con estrellas.
    print(f"    {color} {categoria:15} | {label:8} | Confianza: {score*100:5.2f}% | {'‚≠ê' * estrellas_num}")

print("\n" + "="*90) # L√≠nea divisoria final.
````

````python
# Prepara una lista de diccionarios para crear un DataFrame de pandas con los resultados de la clasificaci√≥n.
datos_clasificacion = []

# Itera a trav√©s de cada frase y su correspondiente resultado de clasificaci√≥n.
for frase, resultado in zip(frases_sentimientos, resultados):
    # Extrae el n√∫mero de estrellas de la etiqueta de clasificaci√≥n.
    estrellas = int(resultado['label'].split()[0])

    # Determina la categor√≠a de sentimiento (Negativo, Neutral, Positivo) bas√°ndose en las estrellas.
    if estrellas <= 2:
        sentimiento = "Negativo"
    elif estrellas == 3:
        sentimiento = "Neutral"
    else:
        sentimiento = "Positivo"

    # A√±ade un diccionario con los datos procesados a la lista `datos_clasificacion`.
    # La frase se trunca si es demasiado larga para la columna 'Frase'.
    datos_clasificacion.append({
        'Frase': frase[:60] + '...' if len(frase) > 60 else frase,
        'Sentimiento': sentimiento,
        'Estrellas': estrellas,
        'Confianza': f"{resultado['score']*100:.1f}%" # Formatea la confianza como porcentaje.
    })

# Crea un DataFrame de pandas a partir de la lista de diccionarios.
# Esto organiza los resultados en una estructura tabular f√°cil de leer y manipular.
df_clasificacion = pd.DataFrame(datos_clasificacion)

# Imprime el DataFrame resultante. `to_string(index=True)` asegura que se muestre
# todo el DataFrame sin truncar, incluyendo el √≠ndice.
print("üìä Tabla de Clasificaci√≥n:\n")
print(df_clasificacion.to_string(index=True))
````

## 6. Generaci√≥n con Modelos Tipo GPT (Decoder)

### ¬øC√≥mo funcionan los modelos generativos?

Los modelos decoder como GPT est√°n dise√±ados para **generar texto** prediciendo el siguiente token de forma autoregresiva.

---

### Proceso de Generaci√≥n

1. **Input**: Prompt inicial ‚Üí `"El gato est√°"`
2. **Predicci√≥n**: Modelo predice siguiente palabra ‚Üí `"sentado"` (probabilidad 60%), `"durmiendo"` (35%), ...
3. **Selecci√≥n**: Se elige una palabra seg√∫n estrategia (temperature, top-k, top-p)
4. **Iteraci√≥n**: Nueva secuencia ‚Üí `"El gato est√° sentado"` ‚Üí predice siguiente ‚Üí ...
5. **Fin**: Hasta alcanzar max_length o token especial de fin

---

### Par√°metros de Generaci√≥n

#### **Temperature (œÑ)**
Controla la "creatividad" del modelo

```
P'(token) = P(token)^(1/œÑ)
```

- **œÑ = 0.1 - 0.5**: Conservador, predecible, determinista
  - Uso: Tareas que requieren precisi√≥n (c√≥digo, datos)
- **œÑ = 0.7 - 1.0**: Balanceado
  - Uso: Escritura general, chatbots
- **œÑ = 1.2 - 2.0**: Creativo, aleatorio, diverso
  - Uso: Escritura creativa, brainstorming

#### **Top-k Sampling**
Limita las opciones a las K palabras m√°s probables

- **k=1**: Greedy decoding (siempre la m√°s probable)
- **k=10-50**: Balanceado
- **k=100+**: M√°s diversidad

#### **Top-p (Nucleus) Sampling**
Considera palabras hasta alcanzar probabilidad acumulada P

- **p=0.5**: Muy conservador
- **p=0.9**: Balanceado (recomendado)
- **p=0.95-1.0**: M√°s diversidad

#### **Max Length / Max New Tokens**
- N√∫mero m√°ximo de tokens a generar

---

### Estrategias de Decodificaci√≥n

| Estrategia | Descripci√≥n | Uso |
|------------|-------------|-----|
| Greedy | Siempre token m√°s probable | Muy determinista, puede ser repetitivo |
| Beam Search | Mantiene N mejores secuencias | Traducciones, res√∫menes |
| Top-k Sampling | Muestrea entre top-K tokens | Balance creatividad/coherencia |
| Top-p Sampling | Probabilidad acumulada | Recomendado para texto general |

````python
# Carga un modelo generativo de texto peque√±o, `distilgpt2`, utilizando el `pipeline` de `transformers`.
# DistilGPT-2 es una versi√≥n m√°s peque√±a y r√°pida de GPT-2, entrenada para tareas de generaci√≥n de texto.
# Nota importante: Este modelo fue entrenado principalmente en datos en ingl√©s.
# Para la generaci√≥n de texto en espa√±ol, se recomendar√≠an modelos espec√≠ficos en espa√±ol (ej., "GPT-2-small-spanish").
generador = pipeline(
    'text-generation', # Especifica la tarea: generaci√≥n de texto.
    model='distilgpt2', # Nombre del modelo pre-entrenado a cargar.
    device=-1 # Ejecuta el modelo en la CPU. Cambiar a 0 para GPU si est√° disponible.
)

# Imprime un mensaje de confirmaci√≥n cuando el modelo generativo ha sido cargado.
print("‚úÖ Modelo generativo cargado (DistilGPT-2)")
````

````python
# Define un prompt inicial (texto de entrada) para la generaci√≥n de texto.
prompt = "Artificial intelligence is transforming"

# Realiza una generaci√≥n de texto b√°sica utilizando el modelo `generador`.
resultado = generador(
    prompt, # El texto de inicio para el modelo.
    max_length=50, # La longitud m√°xima del texto generado (prompt + nueva generaci√≥n).
    num_return_sequences=1, # El n√∫mero de secuencias diferentes a generar.
    do_sample=True, # Activa el muestreo (sampling) para a√±adir variabilidad a la generaci√≥n.
                    # Si fuera `False`, el modelo ser√≠a m√°s determinista (greedy decoding).
    temperature=0.7 # Controla la aleatoriedad de la generaci√≥n. Valores m√°s bajos (ej. 0.2)
                    # hacen el texto m√°s predecible; valores m√°s altos (ej. 1.0) lo hacen m√°s creativo.
)

print("üìù Generaci√≥n B√°sica:\n")
print(f"Prompt: {prompt}") # Imprime el prompt utilizado.
# Imprime el texto generado. `resultado` es una lista de diccionarios, y accedemos
# al primer (y √∫nico) resultado y a la clave 'generated_text'.
print(f"\nTexto generado:\n{resultado[0]['generated_text']}")
````

### üìù EJERCICIO 13: Experimenta con Temperature

````python
# üìù EJERCICIO: Genera textos con diferentes valores de temperature

# Define un prompt inicial para la generaci√≥n de texto.
prompt = "The future of technology will"

# Define una lista de valores de `temperature` para experimentar.
# La temperature controla la aleatoriedad de la generaci√≥n de texto:
# - Baja (ej. 0.2): el modelo elige los tokens m√°s probables, resultando en texto m√°s conservador y predecible.
# - Media (ej. 0.7): equilibrio entre coherencia y creatividad.
# - Alta (ej. 1.2): el modelo es m√°s propenso a elegir tokens menos probables, resultando en texto m√°s creativo,
#                   diverso pero a veces menos coherente.
temperaturas = [0.2, 0.7, 1.2]

print("üå°Ô∏è Experimento: Efecto de la Temperature\n")
print("="*90)

# Itera sobre cada valor de `temperature`.
for temp in temperaturas:
    print(f"\nüîπ Temperature = {temp}")
    print("-"*90)

    # Genera texto utilizando el modelo `generador` con el `prompt` y el valor de `temperature` actual.
    resultados = generador(
        prompt,
        max_length=50, # Longitud m√°xima del texto generado.
        temperature=temp, # Valor de temperature para esta iteraci√≥n.
        num_return_sequences=2,  # Genera 2 versiones de texto para cada `temperature` para comparar.
        do_sample=True # Activa el muestreo para permitir la variabilidad.
    )

    # Imprime cada una de las versiones generadas.
    for i, res in enumerate(resultados, 1):
        print(f"\nVersi√≥n {i}:")
        print(res['generated_text'])

    print("\n" + "-"*90)

print("\nüí° An√°lisis:")
print("   Temperature baja (0.2)  ‚Üí M√°s predecible y coherente")
print("   Temperature media (0.7) ‚Üí Balance entre creatividad y coherencia")
print("   Temperature alta (1.2)  ‚Üí M√°s creativo pero potencialmente incoherente")
````

### üìù EJERCICIO 14: Top-k y Top-p

````python
# üìù EJERCICIO: Compara diferentes configuraciones de top-k y top-p

# Define un prompt inicial para la generaci√≥n de texto.
prompt = "In a distant future, humanity discovered"

# Define diferentes configuraciones para los par√°metros `top_k` y `top_p`.
# `top_k`: El modelo considera solo los `k` tokens m√°s probables para la siguiente predicci√≥n.
#          Un `top_k` bajo hace la generaci√≥n m√°s enfocada; un `top_k` alto permite m√°s diversidad.
# `top_p` (nucleus sampling): El modelo considera el conjunto m√°s peque√±o de tokens
#           cuya probabilidad acumulada supera `p`. `top_p` es a menudo preferido
#           a `top_k` ya que se adapta din√°micamente a la distribuci√≥n de probabilidad.
configuraciones = [
    {"nombre": "Conservador", "top_k": 10, "top_p": 0.5},
    {"nombre": "Balanceado", "top_k": 50, "top_p": 0.9},
    {"nombre": "Creativo", "top_k": 100, "top_p": 0.95}
]

print("üéØ Experimento: Top-k y Top-p Sampling\n")
print("="*90)
print(f"\nPrompt: '{prompt}'\n")

# Itera sobre cada configuraci√≥n definida.
for config in configuraciones:
    print(f"\nüîπ {config['nombre']} (top_k={config['top_k']}, top_p={config['top_p']})")
    print("-"*90)

    # Genera texto con la configuraci√≥n actual de `top_k` y `top_p`.
    # Se mantiene una `temperature` constante (0.8) para aislar el efecto de `top_k`/`top_p`.
    resultado = generador(
        prompt,
        max_length=60, # Longitud m√°xima del texto a generar.
        temperature=0.8, # Un valor de temperature fijo para la comparaci√≥n.
        top_k=config['top_k'], # El n√∫mero de tokens m√°s probables a considerar.
        top_p=config['top_p'], # El umbral de probabilidad acumulada para el muestreo.
        num_return_sequences=1, # Genera una secuencia por configuraci√≥n.
        do_sample=True # Asegura el uso de muestreo para aplicar top_k y top_p.
    )

    # Imprime el texto generado para la configuraci√≥n actual.
    print(f"\n{resultado[0]['generated_text']}")

print("\n" + "="*90)
print("\nüí° Observaciones:")
print("   - top_k bajo: Reduce opciones a las m√°s probables")
print("   - top_p bajo: Corta la 'cola' de probabilidades antes")
print("   - Combinaci√≥n de ambos: Control fino sobre creatividad vs coherencia")
````

## 7. Prompt Engineering B√°sico

### ¬øQu√© es Prompt Engineering?

**Prompt Engineering** es el arte y ciencia de dise√±ar **instrucciones efectivas** para que los modelos de lenguaje generen las respuestas deseadas.

---

### Principios Fundamentales

#### **1. Contexto Claro**
Proporciona informaci√≥n de fondo necesaria

‚ùå Malo: "Explica"
‚úÖ Bueno: "Explica c√≥mo funcionan las redes neuronales en el contexto de visi√≥n por computador"

#### **2. Rol/Persona**
Define qui√©n debe "ser" el modelo

```
"Eres un profesor de f√≠sica explicando a estudiantes de secundaria..."
"Act√∫a como un experto en ciberseguridad..."
```

#### **3. Formato de Salida**
Especifica estructura deseada

```
"Responde en forma de lista numerada"
"Estructura tu respuesta en: 1) Introducci√≥n, 2) Desarrollo, 3) Conclusi√≥n"
"Genera una tabla con columnas: Nombre, Ventaja, Desventaja"
```

#### **4. Ejemplos (Few-shot Learning)**
Muestra el tipo de respuesta esperada

```
"Clasifica el sentimiento:

Texto: 'Me encant√≥ la pel√≠cula'
Sentimiento: Positivo

Texto: 'Fue aburrida'
Sentimiento: Negativo

Texto: 'El servicio fue excelente'
Sentimiento:"
```

#### **5. Restricciones**
L√≠mites de longitud, estilo, contenido

```
"En m√°ximo 3 oraciones..."
"Usando lenguaje t√©cnico apropiado para ingenieros..."
"Sin usar jerga t√©cnica, para audiencia general..."
```

---

### T√©cnicas Avanzadas

#### **Zero-shot**
Sin ejemplos, solo instrucciones
```
"Traduce al franc√©s: ..."
```

#### **One-shot**
Un ejemplo
```
"Escribe en estilo formal:
Informal: '¬øQu√© onda?'
Formal: '¬øC√≥mo est√° usted?'

Informal: '¬øTodo bien?'
Formal:"
```

#### **Few-shot**
M√∫ltiples ejemplos (2-5)

#### **Chain-of-Thought (CoT)**
Pide razonamiento paso a paso
```
"Resuelve paso a paso:
Pregunta: Si 5 manzanas cuestan $10, ¬øcu√°nto cuestan 8 manzanas?

Paso 1:"
```

---

### Errores Comunes

‚ùå **Vagos**: "H√°blame de IA"
‚ùå **Ambiguos**: "Escribe algo sobre esto"
‚ùå **Sin estructura**: Texto largo sin separaciones
‚ùå **Contradictorios**: "S√© breve pero detallado"

---

````python
# Define una funci√≥n auxiliar `generar_respuesta` para simplificar las llamadas al modelo generativo.
# Esta funci√≥n encapsula los par√°metros comunes de generaci√≥n.
def generar_respuesta(prompt, max_len=100, temp=0.7):
    """
    Genera una respuesta de texto utilizando el modelo generativo.

    Args:
        prompt (str): El texto inicial o instrucci√≥n para el modelo.
        max_len (int): La longitud m√°xima del texto generado (incluido el prompt).
        temp (float): La temperature para controlar la aleatoriedad de la generaci√≥n.

    Returns:
        str: El texto generado por el modelo.
    """
    # Llama al pipeline `generador` con los par√°metros especificados.
    resultado = generador(
        prompt, # El prompt de entrada.
        max_length=max_len, # Longitud m√°xima del resultado.
        temperature=temp, # Control de aleatoriedad.
        num_return_sequences=1, # Solo una secuencia de retorno.
        do_sample=True, # Habilita el muestreo.
        pad_token_id=50256  # Especifica el ID del token de relleno, a menudo necesario para
                            # evitar advertencias al usar `max_length` con algunos modelos,
                            # especialmente si el modelo no tiene un token EOS (End Of Sentence) claro.
    )
    # Retorna solo el texto generado de la primera secuencia.
    return resultado[0]['generated_text']

print("‚úÖ Funci√≥n de generaci√≥n definida")
````

### Comparaci√≥n: Prompts Malos vs Buenos

````python
# Este bloque de c√≥digo demuestra la diferencia entre un prompt vago y uno espec√≠fico.
print("üìã Ejemplo 1: Especificidad\n")
print("="*90)

# --- Ejemplo de Prompt MALO (vago) ---
# Un prompt vago no proporciona suficiente contexto, formato o restricciones,
# lo que lleva a respuestas gen√©ricas o no dirigidas.
prompt_malo = "Write about AI"
print(f"\n‚ùå PROMPT MALO (vago):")
print(f"   '{prompt_malo}'\n")
# Genera una respuesta usando el prompt vago.
respuesta_mala = generar_respuesta(prompt_malo, max_len=60)
print(f"Respuesta:\n{respuesta_mala}")

print("\n" + "-"*90 + "\n")

# --- Ejemplo de Prompt BUENO (espec√≠fico y estructurado) ---
# Un prompt bueno es claro, define la audiencia, el formato y las restricciones,
# guiando al modelo a generar una respuesta mucho m√°s √∫til y relevante.
prompt_bueno = """Write a brief explanation of artificial intelligence for high school students.
Include: 1) Definition, 2) One real-world example, 3) One benefit.
Keep it under 100 words."""

print(f"‚úÖ PROMPT BUENO (espec√≠fico y estructurado):")
print(f"   {prompt_bueno}\n")
# Genera una respuesta usando el prompt bien dise√±ado.
respuesta_buena = generar_respuesta(prompt_bueno, max_len=120)
print(f"Respuesta:\n{respuesta_buena}")

print("\n" + "="*90)
````

````python
# Este bloque de c√≥digo ilustra la importancia de proporcionar contexto en los prompts.
print("üìã Ejemplo 2: Importancia del Contexto\n")
print("="*90)

# --- Ejemplo SIN CONTEXTO ---
# Un prompt sin contexto puede ser ambiguo, y el modelo podr√≠a interpretar la pregunta
# de varias maneras, dando una respuesta no deseada.
prompt_sin_contexto = "What is Python?"
print(f"\n‚ùå SIN CONTEXTO:")
print(f"   '{prompt_sin_contexto}'\n")
# Genera una respuesta que podr√≠a no ser la esperada debido a la falta de contexto.
resp1 = generar_respuesta(prompt_sin_contexto, max_len=50)
print(f"Respuesta:\n{resp1}")
print("\n(Podr√≠a referirse al animal o al lenguaje de programaci√≥n)")

print("\n" + "-"*90 + "\n")

# --- Ejemplo CON CONTEXTO CLARO ---
# Al especificar el contexto ("In the context of programming languages"),
# el modelo entiende el dominio de la pregunta y genera una respuesta relevante.
prompt_con_contexto = """In the context of programming languages, explain what Python is
and mention two key features that make it popular."""

print(f"‚úÖ CON CONTEXTO CLARO:")
print(f"   {prompt_con_contexto}\n")
# Genera una respuesta m√°s precisa y √∫til gracias al contexto.
resp2 = generar_respuesta(prompt_con_contexto, max_len=80)
print(f"Respuesta:\n{resp2}")

print("\n" + "="*90)
````
